```python
import pandas as pd
df = pd.read_csv("C:\\Users\\aardr\\Downloads\\Vessel_1_Report_Data.csv")
```

```python
df.head()
```

```python
df['report_date_time'] = pd.to_datetime(df['report_date_time'], format='%d-%m-%Y %H:%M')
```

```python
df['formatted_date'] = df['report_date_time'].dt.strftime('%d-%b-%Y')
df.head()
```

```python
df['Date']= df['report_date_time'].dt.date
df['Time']= df['report_date_time'].dt.time
df['Timetz']= df['report_date_time'].dt.timetz
df['Date']= df['report_date_time'].dt.date
df['Month']= df['report_date_time'].dt.month
df['Day']= df['report_date_time'].dt.day
df['Hour']= df['report_date_time'].dt.hour
df['Minute']= df['report_date_time'].dt.minute
df['Second']= df['report_date_time'].dt.second
df['Microsecond']= df['report_date_time'].dt.microsecond
df['Nanosecond']= df['report_date_time'].dt.nanosecond
df['Dayofweek']= df['report_date_time'].dt.dayofweek
df['Day_of_week']= df['report_date_time'].dt.day_of_week
df['Dayofweek']= df['report_date_time'].dt.dayofweek
df['weekday']= df['report_date_time'].dt.weekday
df['dayofyear']= df['report_date_time'].dt.dayofyear
df['days_in_month']= df['report_date_time'].dt.days_in_month
df['quarter']= df['report_date_time'].dt.quarter
df['month_start']= df['report_date_time'].dt.is_month_start
df['month_end']= df['report_date_time'].dt.is_month_end
df['quarter_start']= df['report_date_time'].dt.is_quarter_start
df['quarter_end']= df['report_date_time'].dt.is_quarter_end
df['is_year_start']= df['report_date_time'].dt.is_year_start
df['is_year_end']= df['report_date_time'].dt.is_year_end
df['is_leap_year']= df['report_date_time'].dt.is_leap_year
df['daysinmonth']= df['report_date_time'].dt.daysinmonth
df['days_in_month']= df['report_date_time'].dt.days_in_month
df['period']= df['report_date_time'].dt.to_period('M')
df['strftime']= df['report_date_time'].dt.strftime('%a')
df['Round']= df['report_date_time'].dt.round(freq='h')
df['Floor']= df['report_date_time'].dt.floor('h')
df['Ceil']= df['report_date_time'].dt.ceil('h')
df['Month_Name']= df['report_date_time'].dt.month_name()
df['Day_Name']= df['report_date_time'].dt.day_name()
df.head()
```

```python
df.to_csv('Aardra_Vessel.csv', index=False)
print("Saved")
```
